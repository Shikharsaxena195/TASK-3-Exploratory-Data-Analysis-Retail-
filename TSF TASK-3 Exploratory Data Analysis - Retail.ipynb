```python
# importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset

data = pd.read_csv("C:\\Users\\KIIT\\Desktop\\SampleSuperstore.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
#getting the statistical information of dataset

data.describe()
```

```python
#checking in the category field

data.groupby(['Category']).mean()
```

```python
data.groupby(['Category']).max()
```

```python
data.groupby(['Category']).min()
```

```python
#visualizing the total profit as compared to sales 

plt.figure(figsize= (10,20))

data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()

plt.show()
```

```python
#checking in sub-category field

data.groupby('Sub-Category').mean()
```

```python
data.groupby('Sub-Category').max()
```

```python
data.groupby('Sub-Category').min()
```

```python
#for sub category we are plotting bar plot

plt.figure(figsize= (15,25))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot(kind='bar')
plt.show()
```

# Corelation

```python
corr = data.corr()
corr
```

# Heat Map

```python
sns.heatmap(corr, annot=True)
plt.show()
```

```python
#Sales and Profit are highly co related(0.48)
#More the Sales, more the Profit. 
```

```python
#Ploting based upon the REGION and the category

plt.figure(figsize=(20,6))
plt.subplot(2,1,1)
sns.countplot("Region", hue="Category", data=data)
plt.title("Region-Category Barplot")
plt.xlabel("REGION", fontsize=12)
plt.ylabel("COUNT", fontsize=12)

plt.grid(alpha=0.3)
plt.show()
```

```python
#Ploting based upon the REGION and the sub-category

plt.figure(figsize=(20,6))

plt.subplot(2,1,2)

sns.countplot('Region',hue='Sub-Category',data=data)

plt.title('REGIONS-SUB CATEGORY')
plt.ylabel('COUNT',fontsize=12)
plt.xlabel('REGION',fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
#Now the Analysis on the basis of shiping mode

sns.countplot(x=data['Ship Mode'])
plt.show()
```

```python
#Standard class has the highest number of shipping mode
```

```python
#Now Pie-chart for sub-category

plt.figure(figsize=(10,15))

data['Sub-Category'].value_counts().plot(kind='pie', autopct="%1.1f%%")
plt.show()
```

```python
#Binders having the highest value count(15.2%)
```

```python
#Sales per Ship mode

data.groupby('Ship Mode')['Sales','Profit'].agg(['sum']).plot(kind='bar')

plt.title('Year wise Total Sales & % of profit gained')
```

```python
#Checking for the state wise number of delivery.

plt.figure(figsize=(13,8))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
#California state has the highest number of delivery. 
```

```python
most_profit_state = data.groupby("State").sum().sort_values('Profit', ascending=False)
most_profit_state.head(12)
```

```python
#California is the state with maximum Profit. 
```

```python
most_sales_state = data.groupby("State").sum().sort_values('Sales', ascending=False)
most_sales_state.head(12)
```

```python
#California is the state with maximum Sales. 
```

```python
#Pairplot betweeen Sub-Category and different parametres

figsize=(15,10)
sns.pairplot(data,hue='Sub-Category')
plt.show()
```

```python
#Pairplot betweeen Category and different parametres

figsize=(15,10)
sns.pairplot(data,hue='Category')
plt.show()
```
